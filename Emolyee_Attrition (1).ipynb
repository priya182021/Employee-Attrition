```python
y = df['Attrition']
x = df.drop(['Attrition'], axis = 1)
```

```python
# lets check the shape of x and y
print("Shape of x :", x.shape)
print("Shape of y :", y.shape)
```

Resampling for Balancing the Data

```python
import imblearn
```

```python
# It is very important to resample the data, as the Target class is Highly imbalanced.
# Here We are going to use Over Sampling Technique to resample the data.
# lets import the SMOTE algorithm to do the same.

from imblearn.over_sampling import SMOTE

xr, yr = SMOTE(k_neighbors=3).fit_resample(x, y)

# lets print the shape of x and y after resampling it
print(xr.shape)
print(yr.shape)
```

```python
# lets also check the value counts of our target variable4

print("Before Resampling :")
print(y.value_counts())

print("After Resampling :")
yr = pd.DataFrame(yr)
print(yr.value_counts())
```

```python
# lets split the test data from the training data

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(xr, yr, test_size = 0.2, random_state = 0)

# lets print the shapes again 
print("Shape of the x Train :", x_train.shape)
print("Shape of the y Train :", y_train.shape)
print("Shape of the x Test :", x_test.shape)
print("Shape of the y Test :", y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier 

model = RandomForestClassifier()
model.fit(x_train , y_train)
```

```python
print("Testing Accuracy : " , model.score(x_test , y_test))
```

```python
predictions = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix 

matrix = confusion_matrix(y_test , predictions)
matrix
```

```python
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score

#Accuracy and Classification report
print(accuracy_score(y_test,predictions))
print(classification_report(y_test, predictions))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier 
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
#initalizing the Gradientboost model
classifier = GradientBoostingClassifier (min_samples_split=2,max_depth=3,max_features=None,learning_rate=0.01 , n_estimators = 50,random_state=0)
classifier.fit(x_train, y_train)
```
