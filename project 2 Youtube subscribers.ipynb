```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("topSubscribed.csv")
```

```python
df.set_index("Rank",inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Subscribers"]=df["Subscribers"].str.replace(',','')
df["Video Views"]=df["Video Views"].str.replace(',','')
df["Video Count"]=df["Video Count"].str.replace(',','')
```

```python
df["Subscribers"]=df["Subscribers"].astype("float64")
df["Video Views"]=df["Video Views"].astype("float64")
df["Video Count"]=df["Video Count"].astype("float64")
```

```python
df.columns=df.columns.str.replace(' ','_')
```

```python
df["Total_Views"]=df["Video_Views"]-df["Video_Count"]
```

```python
df["Total_Views"]=df["Total_Views"]/1000000
```

```python
df[["Total_Views","Youtube_Channel"]].sort_values(by = "Total_Views", ascending  = False).head(10)
```

```python
df["Category"].value_counts()
```

```python
sns.barplot(y="Started", x="Youtube_Channel",data=df, palette = 'magma')
plt.xticks(rotation = 90, fontsize = 14)
plt.rcParams['figure.figsize'] = (20, 15)
plt.style.use('fivethirtyeight')
plt.xlabel(' ')
plt.show()
```

```python
sns.countplot(x="Started",data=df)
plt.figure(figsize=(14,12))
plt.show()
```
